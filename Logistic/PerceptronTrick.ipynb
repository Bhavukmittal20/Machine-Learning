```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self,epochs=1000,eta=0.01):
        self.epochs=epochs
        self.eta=eta
        self.coef_=np.array([])
    def fit(self,X_train,y_train):
        X_train=np.hstack((np.ones((X_train.shape[0],1)),X_train))
        self.coef_=np.ones(X_train.shape[1])
        for i in range(self.epochs):
            a=np.random.randint(X_train.shape[0])
            y_hat=self.sigmoid(np.dot(X_train[a],self.coef_.T))
            self.coef_=self.coef_+(y_train[a]-y_hat)*self.eta*(X_train[a])
    def predict(self,X_test):
        X_test=np.hstack((np.ones(X_test.shape[0],1),X_test))
        return np.dot(X_test,self.coef_)>0
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))

```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,class_sep=10,hypercube=False,random_state=41)
plt.scatter(X[:,0],X[ :,1],c=y,cmap='winter')
plt.show()
```

```python
a=Perceptron(eta=0.01,epochs=100000)
a.fit(X,y)
b = a.coef_[0]
w1 = a.coef_[1]
w2 = a.coef_[2]
m = -(w1 / w2)
c = -(b / w2)
x_input = np.linspace(X[:,0].min(), X[:,0].max(), 100)
y_output = m * x_input + c
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter') 
plt.plot(x_input, y_output, color='red', linewidth=3) 
plt.ylim(X[:,1].min()-1, X[:,1].max()+1)
plt.show()
```
