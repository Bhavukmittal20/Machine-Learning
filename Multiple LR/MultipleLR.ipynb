Formula Used:

Coefficeients=((X^T.X)^-1).X^T.Y
Where X matrix is 1 column of 1 + all the training examples and y is the output

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class LRegress:
    def __init__ (self):
        self.coef_=np.array([])
    def fit(self,X_train,y_train):
        X_t=np.hstack([np.ones((X_train.shape[0],1)),X_train])
        self.coeff=np.linalg.inv(np.dot(X_t.T,X_t)).dot(X_t.T).dot(y_train)
    def predict(self,X_test):
        X_t=np.hstack([np.ones((X_test.shape[0],1)),X_test])
        return np.dot(X_t,self.coeff)
```

```python
from sklearn.datasets import load_diabetes
X,y=load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
reg.predict(X_test)
```

```python
mylr=LRegress()
mylr.fit(X_train,y_train)
mylr.predict(X_test)
```

```python
print(mylr.coeff)
print(reg.intercept_,reg.coef_)
```

Hence all coeffients and intercepts are same hence we correctly executed the Linear Regression Algorithm
